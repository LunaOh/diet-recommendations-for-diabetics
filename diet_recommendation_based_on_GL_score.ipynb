```python
import pandas as pd
import numpy as np

df = pd.read_excel('/content/drive/MyDrive/혼자한/전국통합식품영양성분정보(음식)표준데이터-20230510.xls')
df
```

```python
df.columns = df.iloc[0]
df = df[1:]
```

```python
selected_columns = ['식품명', '식품대분류명', '에너지(kcal)', '단백질(g)', '지방(g)', '탄수화물(g)']
subset_df = df[selected_columns]

subset_df.dropna(inplace=True)
```

```python
subset_df.info()
```

```python
# 숫자 열을 float 형식으로 변환
numeric_columns = ['에너지(kcal)', '단백질(g)', '지방(g)', '탄수화물(g)']
subset_df[numeric_columns] = df[numeric_columns].astype(float)

print(subset_df.info())
```

```python
# GI지수와 GL지수 계산
def calculate_gi(glucose_value, carb_value):
    return round((carb_value / glucose_value) * 100, 2)

def calculate_gl(gi_value, carb_value):
    return round((gi_value * carb_value) / 100, 2)

subset_df['GI지수'] = np.vectorize(calculate_gi)(100, subset_df['탄수화물(g)'])
subset_df['GL지수'] = np.vectorize(calculate_gl)(subset_df['GI지수'], subset_df['탄수화물(g)'])

print(subset_df.head())
```

```python
subset_df['식품대분류명'].unique()
```

```python
subset_df['식품대분류명'] = subset_df['식품대분류명'].str.replace('·', ',')
```

```python
subset_df
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# 선호도 입력
fat_preference = float(input("지방 선호도(0~1): "))
carb_preference = float(input("탄수화물 선호도(0~1): "))
protein_preference = float(input("단백질 선호도(0~1): "))
gl_preference = float(input("GL지수 선호도(0~1): "))

subset_df = subset_df.reset_index(drop=True)

# 식품특성 벡터화
features = np.array([[food["GL지수"], food["지방(g)"], food["탄수화물(g)"], food["단백질(g)"]] for _, food in subset_df.iterrows()])

# 선호도 벡터화
preference_vector = np.array([gl_preference, fat_preference, carb_preference, protein_preference])

# 코사인 유사도 계산
similarity_scores = cosine_similarity(features, [preference_vector])

# 필터링
mask = subset_df["GL지수"] <= 10
filtered_scores = similarity_scores.flatten() * mask

recommendations = [subset_df.iloc[i]["식품명"] for i in np.argsort(filtered_scores)[::-1] if filtered_scores[i] > 0]

top_recommendations = recommendations[:5]

print("식품 추천 결과:")
if top_recommendations:
    for recommendation in top_recommendations:
        print(recommendation)
else:
    print("추천할 식품이 없습니다.")
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# 선호도 입력
category_preference = input("식품 대분류명: ")

# 식품특성 벡터화
features = np.array([[food["GL지수"], food["지방(g)"], food["탄수화물(g)"], food["단백질(g)"]] for _, food in subset_df.iterrows()])

# 코사인 유사도 계산
similarity_scores = cosine_similarity(features, features)

# 필터링
mask = (subset_df["GL지수"] <= 10) & (subset_df["식품대분류명"] == category_preference)

filtered_recommendations = subset_df.loc[mask].copy()
filtered_recommendations["유사도"] = similarity_scores[:, 0][mask]
filtered_recommendations.sort_values(by=["GL지수", "유사도"], inplace=True)

top_recommendations = filtered_recommendations.iloc[:5]["식품명"].tolist()

print("식품 추천 결과:")
if top_recommendations:
    for recommendation in top_recommendations:
        print(recommendation)
else:
    print("추천할 식품이 없습니다.")
```
